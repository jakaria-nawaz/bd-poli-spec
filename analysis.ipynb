```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_parquet("data/responses.parquet")
print(f"{len(df)} responses loaded")
print(df.dtypes)
```

```python
print(f"Total responses: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print("\nAll columns:")
for column_name in sorted(df.columns):
    print(f"  {column_name}")

print("Null counts per column:")
print(df.isnull().sum().sort_values(ascending=False))

# See what one full row looks like.
pd.set_option("display.max_rows", 100)
df.iloc[0]
```

```python
# Score distributions
print(df["economicScore"].describe())
print(df["socialScore"].describe())

# Result label breakdown
print(df["resultLabel"].value_counts())

# Demographics
print(df["gender"].value_counts())
print(df["occupation"].value_counts())
print(df["education"].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df["economicScore"].hist(bins=30, ax=axes[0], color="steelblue")
axes[0].set_title("Economic Score Distribution")
df["socialScore"].hist(bins=30, ax=axes[1], color="coral")
axes[1].set_title("Social Score Distribution")
plt.tight_layout()
plt.show()
```

```python
pd.crosstab(df["occupation"], df["resultLabel"])
```

```python
# Gender-based classification across the political spectrum.
gender_spectrum_counts = pd.crosstab(df["gender"], df["resultLabel"])
print(gender_spectrum_counts)

print("\nPercentage within each gender:")
gender_spectrum_share = pd.crosstab(
    df["gender"],
    df["resultLabel"],
    normalize="index",
).mul(100).round(1)
print(gender_spectrum_share)

print("\nAverage scores by gender:")
print(df.groupby("gender")[["economicScore", "socialScore"]].mean().round(2))
```

```python
# Metadata coverage and location hints.
location_columns = ["meta_country", "meta_region", "meta_city"]
print("Non-null counts for location-related metadata:")
print(df[location_columns].notna().sum())

print("\nTop countries:")
print(df["meta_country"].value_counts(dropna=False).head(15))

print("\nTop regions:")
print(df["meta_region"].value_counts(dropna=False).head(15))

print("\nTop cities:")
print(df["meta_city"].value_counts(dropna=False).head(15))

print("\nLocation note:")
print("Most rows do not include country, region, or city, so localisation is only a partial signal.")
```

```python
# Country-based political spectrum using percentages.
top_countries = df["meta_country"].value_counts().head(10).index
country_subset = df[df["meta_country"].isin(top_countries)].copy()

country_spectrum_counts = pd.crosstab(
    country_subset["meta_country"],
    country_subset["resultLabel"],
)
print("Counts for top countries:")
print(country_spectrum_counts)

country_spectrum_percent = pd.crosstab(
    country_subset["meta_country"],
    country_subset["resultLabel"],
    normalize="index",
).mul(100).round(1)
print("\nPercentage within each country:")
print(country_spectrum_percent)

print("\nRows included in this country analysis:")
print(country_subset["meta_country"].value_counts())
```
